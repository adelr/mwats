```python
import requests
import pandas as pd
import numpy as np
```

```python
interesting=pd.read_excel('interesting_sources.xlsx')
interesting.head()
```

```python
def query(ra, dec):
    url = f"http://simbad.u-strasbg.fr/simbad/sim-coo?Coord={ra}d{dec}d&CooFrame=FK5&CooEpoch=2000&CooEqui=2000&CooDefinedFrames=none&Radius=2&Radius.unit=arcmin&submit=submit+query"
    return url
```

```python
def get_response(ra, dec):

    response = requests.get(query(ra, dec))

    tables = pd.read_html(response.text)
    
    idx = np.argmax(['Identifier' in t.columns for t in tables])

    table = tables[idx]
    
    try:
        
        identifier = table.loc[0, 'Identifier']
        
    except KeyError:
        identifier = ''

    return identifier, table
```

```python
ra, dec = interesting.loc[0, 'ra'], interesting.loc[0, 'dec']
i, t = get_response(ra, dec)
print(i)
t
```

```python
interesting.iloc[:10, :].apply(lambda source: get_response(source.ra, source.dec)[0], axis=1)
```
